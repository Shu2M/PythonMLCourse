### Разделение данных

К наборам данных с явным дисбалансом (коим и является наш датасет) применяют методы перебалансировки, в частности метод недосемплирования (Undersampling the majority class). Его идея состоит в том, чтобы большой класс заменить подвыборкой, по мощности равной малому классу. Для пееребалансировки данных воспользуемся методом NearMiss-1, реализованным в библиотеке imblearn

```python
from imblearn.under_sampling import NearMiss
```

```python
X = credit_card_df.drop('Class', axis=1)
y = credit_card_df['Class']
```

```python
nm1 = NearMiss(version=1)
X_b, y_b = nm1.fit_resample(X, y)
```

Посмотрим на количества вхождений первого и второго класса в сбалансированной выборке

```python
y_b.value_counts()
```

Сравним статистики получившихся подвыборок и генеральной выборки

```python
X_b.describe()
```

```python
y_b.describe()
```

При сравнении становится видно, что существенно изменились показатели std признаков - они стали больше по сравнению с генеральной выборкой

Снова построим карту распределения корреляций и выберем важные признаки для определения мошенников

```python
credit_card_balanced_df = pd.concat([X_b, y_b], axis=1)

plt.figure(figsize=(30,20))
sns.set(font_scale=1.15)
sns.heatmap(
    credit_card_balanced_df.corr(),
    annot=True,
    vmin=0, 
    vmax=1,
    mask=np.triu(credit_card_balanced_df.corr()),
)
```

Теперь признаки, влияющие на класс отчетливо видны и их можно выбрать

```python
X_b = X_b[[
    'V2',
    'V4',
    'V8',
    'V11',
    'V19',
    'V20',
    'V21',
    'V27',
    'Amount'
]]
```

Для дальнейшего обучения моделей проведем деление новых подвыборок на тестовую и обучающую. Для этого воспользуемся функцией train_test_split модуля sklearn. Так же предварительно нормализуем данные с помощью инстремента StandardScaler той же библиотеки

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_b_normalized = scaler.fit_transform(X_b)
```

```python
X_b_normalized_train, X_b_normalized_test, y_b_train, y_b_test = train_test_split(
    X_b_normalized, 
    y_b, 
    random_state=0, 
    test_size=0.2,
)
```

### Обучение моделей на сбалансированной выборке

Для предсказания воспользуемся моделями KNeighborsClassifier, LogisticRegression и SVC. В качестве метрики точности моделей будем использовать precision, recall и f1 меру. Ниже представленны формулы этих метрик:

$ precision = \frac{TP}{TP + FP} $

$ recall = \frac{TP}{TP + FN} $

$ f_1 = 2\frac{precision * recall}{precision + recall} $

где

$ TP $ - True Positive, положительное предсказание соответсвует положительной  истине

$ FP $ - False Positive, положительное предсказание не соответсвует отрицательной истине (ошибка I рода)

$ FN $ - False Negative, отрицательное предсказание не соответсвует положительной истине (ошибка II рода)

Так, метрика precision характеризует долю объектов, названных положительными и при этом действительно являющихся положительными. Иными словами, это способность модели отличать класс от других. Метрика recall показывает, какую долю объектов положительного класса из всех объектов положительного класса нашел алгоритм, то есть способность алгоритма обнаруживать класс вообще. f1 метрика - это среднее гармоническое этих двух метрик. 
В случае нашей задачи, когда положительных классов (мошенников) мало по сравнению с остальным количеством данных, стоит обращать внимание именно на метрику recall, так как она содержит ошибку II рода, что в данной задаче критично (мы не хотим принимать мошенников за обычных пользователей). Что же до метрики precision, то она в частности является показателем количества накопленых ошибок I рода, так как содержит FP, однако в данной задаче, это не столь критично, как ошибки первого рода (лучше признать добросовестного человека мошенником и лишний раз перепроверить его, чем наоборот).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

#### 1) KNeighborsClassifier

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_b_normalized_train, y_b_train);
```

```python
knn_y_pred = knn_model.predict(X_b_normalized_test)
```

```python
print(classification_report(y_b_test, knn_y_pred))
```

#### 2) LogisticRegression

```python
logreg_model = LogisticRegression(random_state=42, max_iter=10000)
logreg_model.fit(X_b_normalized_train, y_b_train);
```

```python
logreg_y_pred = logreg_model.predict(X_b_normalized_test)
```

```python
print(classification_report(y_b_test, logreg_y_pred))
```

#### 3) SVC

```python
svc_model = SVC(random_state=42)
svc_model.fit(X_b_normalized_train, y_b_train);
```

```python
svc_y_pred = svc_model.predict(X_b_normalized_test)
```

```python
print(classification_report(y_b_test, svc_y_pred))
```

Высокие показатели метрик recall и precision указывают на малое количество ошибок первого и второго рода, что говорит о хорошей способности алгоритмов различать мошеннические транзакции и обычные

### Обучение моделей на несбалансированной выборке

Сравним полученные результаты метрик с метриками моделей, обученных на несбалансированных данных

```python
X_unb = X[[
    'V2',
    'V4',
    'V8',
    'V11',
    'V19',
    'V20',
    'V21',
    'V27',
    'Amount'
]]

X_unb_normalized = scaler.fit_transform(X_unb)

X_unb_normalized_train, X_unb_normalized_test, y_unb_train, y_unb_test = train_test_split(
    X_unb_normalized, 
    y, 
    random_state=0, 
    test_size=0.2,
)
```

#### 1) KNeighborsClassifier

```python
knn_model_unb = KNeighborsClassifier()
knn_model_unb.fit(X_unb_normalized_train, y_unb_train);
```

```python
knn_y_pred_unb = knn_model_unb.predict(X_unb_normalized_test)
```

```python
print(classification_report(y_unb_test, knn_y_pred_unb))
```

#### 2) LogisticRegression

```python
logreg_model_unb = LogisticRegression(random_state=42, max_iter=10000)
logreg_model_unb.fit(X_unb_normalized_train, y_unb_train);
```

```python
logreg_y_pred_unb = logreg_model_unb.predict(X_unb_normalized_test)
```

```python
print(classification_report(y_unb_test, logreg_y_pred_unb))
```

#### 3) SVC

```python
svc_model_unb = SVC(random_state=42)
svc_model_unb.fit(X_unb_normalized_train, y_unb_train);
```

```python
svc_y_pred_unb = svc_model_unb.predict(X_unb_normalized_test)
```

```python
print(classification_report(y_unb_test, svc_y_pred_unb))
```

Как можно видеть, precision в целом осталась на довольно высоком уровне, что говорит о способности модели отличать класс мошеннической транзации от обычной, и допускать малое количество ошибок первого рода. Напротив, метрика recall сильно просела, что говорит о большом количестве ошибок второго рода, критичный для данной задачи - такой алгоритм просто разорит компанию!

### Заключение
